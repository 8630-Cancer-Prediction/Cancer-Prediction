```python
import pandas
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
```

```python
Path_list=['/content/drive/MyDrive/Untitled folder/output_pso_feature_reduction.csv','/content/drive/MyDrive/Untitled folder/Preprocessed_BRCA_RNA_data_normalized.csv',
           '/content/drive/MyDrive/Untitled folder/Preprocessed_Data_With_ClassEncoded.csv']


for k in range(len(Path_list)):
  df = pandas.read_csv(Path_list[k])
  print(Path_list[k])

  train_list= df.columns.drop("Class").drop("Sample-id")
  X = df[train_list]
  y = df['Class']

  NaiveBayes = GaussianNB()
  MLP = MLPClassifier(hidden_layer_sizes=250, activation='relu', 
                      solver='adam', alpha=0.0001, batch_size=500, learning_rate='constant', 
                      learning_rate_init=0.001, shuffle=True, random_state=None)
  RandomForest = RandomForestClassifier(max_depth=None, random_state=0)

  model_classifiers=[NaiveBayes, MLP, RandomForest]
  #model_classifiers=[RandomForest]
  

  for i in range(len(model_classifiers)):
    global pred_dict
    pred_dict=[]

    def cscorer(clf, X, y):
        y_pred = clf.predict(X)
        metric = metrics.classification_report(y,y_pred)
        pred_dict.append(metric)
        return 1

    all_pred = cross_val_score(model_classifiers[i], X, y, cv=5, scoring= cscorer)

    for j in pred_dict:
      print("The classification reports of Classfier with 5-fold cross-validation")
      print (j)  
```

```python
with open('/content/drive/MyDrive/Untitled folder/classification_report.csv', 'w') as wr_file:
  for j in range(5):
    div = pred_dict[j].split('\n')
    for i, item in enumerate(div):
      line_split = item.split()
      if (i>1 and i<5):
        for word in line_split:
          wr_file.write(word+',')
        wr_file.write('\n')
      
      if (i == 6):
        wr_file.write('accuracy, , ,')
        wr_file.write(line_split[1] + ','+line_split[2] + '\n')
      
      if (i>6 and i<9):
        for word in line_split:
          if word != 'avg':
            wr_file.write(word+',')
        wr_file.write('\n')
    wr_file.write('\n')
```
